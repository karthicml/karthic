```python
def gameboard(x):
    print(' {} | {} | {} ' .format(x[0],x[1],x[2]))
    print('-----------')
    print(' {} | {} | {} ' .format(x[3],x[4],x[5]))
    print('-----------')
    print(' {} | {} | {} ' .format(x[6],x[7],x[8]))
```

```python
def game():
    print('Welcome to Tic Tac Toe game')
    print(' ')
    a = input('Player 1: Enter your name - ')
    b = input('Player 2: Enter your name - ')
    print(' ')
    gamesetup(a,b)

def gamesetup(a,b):
    x = ['1','2','3','4','5','6','7','8','9']
    y = x
    gameboard(x)
    print(' ')
    i = 0
    while gamelogic(x,a,b) == False and i <9:
        one = int(player1(a,y))
        while one == False:
            one = int(player1(a,y))
        x[one-1] = 'O'
        y = x
        gameboard(x) 
        print(' ')
        i+=1
        if gamelogic(x,a,b) == False and i<9:
            two = int(player2(b,y))
            while two == False:
                two = int(player2(b,y))
            x[two-1] = 'X'
            y = x
            gameboard(x)
            print(' ')
            i+=1
        elif gamelogic(x,a,b) == False and i>8:
            print('Draw!')
            return
        else:
            pass 
    winner = gamelogic(x,a,b)
    print(' ')
    print('Congrats {}! You Win!!' .format(winner[1]))
    print(' ')
    playagain(a,b)
```

```python
def player1(a,y):
    valid1 = False
    while valid1 == False:
        in1 = input('{}: Choose your position - ' .format(a))
        valid1 = validinput(in1)
    one = int(in1)
    print(' ')
    if y[one-1] == 'O' or y[one-1] == 'X':
        print('Selected position is already taken, enter a valid position')
        return False
    else:
        return y[one-1]
    print(' ')

def player2(b,y):
    valid2 = False
    while valid2 == False:
        in2 = input('{}: Choose your position - ' .format(b))
        valid2 = validinput(in2)
    two = int(in2)
    print(' ')
    if y[two-1] == 'O' or y[two-1] == 'X':
        print('Selected position is already taken, enter a valid position')
        return False
    else:
        return y[two-1]
    print(' ')

def gamelogic(x,a,b):
    for i in range (0,3):
        if x[i] == x[i+3] and x[i] == x[i+6]:
            k = i
            winner(x,k,a,b)
            return (True,winner(x,k,a,b))
    for j in range (0,8,3):
        if x[j] == x[j+1] and x[j] == x[j+2]:
            k = j
            winner(x,k,a,b)
            return (True,winner(x,k,a,b))
    else:
        if x[0] == x[4] and x[0] == x[8]:
            k = 0
            winner(x,k,a,b)
            return (True,winner(x,k,a,b))
        elif x[2] == x[4] and x[2] == x[6]:
            k = 2
            winner(x,k,a,b)
            return (True,winner(x,k,a,b))
        else:
            return False
            
def winner(x,k,a,b):
    if x[k] == 'O':
        return a
    else:
        return b
    
def validinput(x):
    if x.isdigit() == False:
        print('')
        print('Your input is not a valid digit. Enter a valid digit.')
        print('')
        return False
    elif int(x) <1 or int(x) > 9:
        print('')
        print('Your input is not in valid range. Enter a digit between 1-9.')
        print('')
        return False
    else:
        return True

def playagain(a,b):
    x = input('Want to play again (Y/N): ')
    print(' ')
    if x == 'Y' or x =='y':
        gamesetup(a,b)
        print('')
    elif x == 'N' or x == 'n':
        pass
    else:
        print('Choose either Y or N')
        playagain()
```
